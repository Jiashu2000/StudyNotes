## Exercise 1: Chipo

### 1. Import Libraries

```python
import pandas as pd
import numpy as np
```

### 2. Import Dataset

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

```python
chipo = pd.read_csv(url, sep = "\t")
```

```python
chipo.head()
```

### 3. Product Cost More Than 10.00

```python
chipo.dtypes
```

```python
chipo['price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

```python
chipo.dtypes
```

```python
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity', 'choice_description'])
```

```python
chipo_filtered.loc[(chipo_filtered['price'] > 10) & (chipo_filtered['quantity'] == 1)]
```

### 4. Price of Each Item

```python
chipo_filtered.loc[chipo_filtered['quantity'] == 1][['item_name', 'price']]
```

### 5. Sort Item Name

```python
chipo.sort_values(by = 'item_name')
```

### 6. Quantity of the Most Expensive Item Ordered

```python
chipo.sort_values(by = 'price', ascending = False)
```

```python
chipo.sort_values(by = 'price', ascending = False).iloc[0]['quantity']
```

### 7. The number of Times Veggie Salad Bowl was Ordered

```python
chipo.loc[chipo['item_name'] ==  'Veggie Salad Bowl']
```

```python
chipo.loc[chipo['item_name'] ==  'Veggie Salad Bowl']['order_id'].nunique()
```
