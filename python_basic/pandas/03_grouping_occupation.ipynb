```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

```python
users = pd.read_csv(url, sep = '|')
```

```python
users.head()
```

```python
users.groupby('occupation')['age'].mean()
```

```python
d = users.groupby(['occupation', 'gender'])[['user_id']].count()
```

```python
d
```

```python
g = d.unstack()
```

```python
g
```

```python
g.columns
```

```python
g.fillna(0, inplace = True)
```

```python
(g['user_id', 'M'] / (g['user_id', 'M'] + g['user_id', 'F'])).sort_values(ascending = False)
```

```python
users.groupby('occupation')['age'].agg(['min', 'max'])
```

```python
g = users.groupby(["occupation", 'gender']).size().unstack()
```

```python
g.fillna(0, inplace = True)
```

```python
g['T'] = g['F'] + g['M']
```

```python
g
```

```python
g['F'] = g['F']/ g['T']
g['M'] = g['M']/ g['T']
```
